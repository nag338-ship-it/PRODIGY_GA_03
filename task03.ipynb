```python
import random
import re
from collections import defaultdict

print("✅ Libraries imported!")
```

```python
# Training text — the more text the better!
training_text = """
Artificial intelligence is transforming the world in many ways.
Machine learning is a subset of artificial intelligence.
Deep learning is a subset of machine learning that uses neural networks.
Neural networks are inspired by the human brain structure.
The future of artificial intelligence is very promising.
Generative AI can create text images music and even code.
Natural language processing helps computers understand human language.
Computer vision allows machines to interpret and understand images.
Reinforcement learning allows agents to learn through trial and error.
Data science combines statistics mathematics and computer science.
Python is the most popular programming language for machine learning.
TensorFlow and PyTorch are popular deep learning frameworks.
Hugging Face provides thousands of pre-trained AI models.
GPT is a large language model developed by OpenAI.
Stable Diffusion is a powerful image generation model.
The transformer architecture revolutionized natural language processing.
Attention mechanisms help models focus on relevant parts of input.
Transfer learning allows models to apply knowledge from one task to another.
Generative adversarial networks consist of a generator and discriminator.
The discriminator tries to distinguish real from generated samples.
"""

print("✅ Training text ready!")
print(f"Total characters: {len(training_text)}")
print(f"Total words: {len(training_text.split())}")
```

```python
def build_markov_chain(text, order=2):
    """Build a Markov chain from text"""
    words = text.lower().split()
    markov_chain = defaultdict(list)

    for i in range(len(words) - order):
        # Create state from 'order' number of words
        state = tuple(words[i:i+order])
        next_word = words[i+order]
        markov_chain[state].append(next_word)

    print(f"✅ Markov Chain built!")
    print(f"Total states: {len(markov_chain)}")
    return markov_chain

# Build chain with order 2 (bigram)
markov_chain = build_markov_chain(training_text, order=2)
```

```python
def generate_text(markov_chain, seed_words, num_words=50, order=2):
    """Generate text using Markov chain"""

    # Start with seed words
    current_state = tuple(seed_words[-order:])
    result = list(seed_words)

    for _ in range(num_words):
        if current_state in markov_chain:
            next_word = random.choice(markov_chain[current_state])
            result.append(next_word)
            current_state = tuple(result[-order:])
        else:
            break

    return ' '.join(result)

print("✅ Text generation function ready!")
```

```python
# Test with different seed words
test_seeds = [
    ["artificial", "intelligence"],
    ["machine", "learning"],
    ["deep", "learning"],
    ["natural", "language"],
    ["generative", "ai"]
]

print("=" * 60)
print("🤖 MARKOV CHAIN TEXT GENERATION RESULTS")
print("=" * 60)

for seed in test_seeds:
    generated = generate_text(markov_chain, seed, num_words=30)
    print(f"\n📌 Seed: '{' '.join(seed)}'")
    print(f"📝 Generated: {generated}")
    print("-" * 60)
```

```python
def build_char_markov(text, order=3):
    """Build character level Markov chain"""
    char_chain = defaultdict(list)

    for i in range(len(text) - order):
        state = text[i:i+order]
        next_char = text[i+order]
        char_chain[state].append(next_char)

    return char_chain

def generate_char_text(char_chain, seed, num_chars=200, order=3):
    """Generate text character by character"""
    result = seed
    current_state = seed[-order:]

    for _ in range(num_chars):
        if current_state in char_chain:
            next_char = random.choice(char_chain[current_state])
            result += next_char
            current_state = result[-order:]
        else:
            break

    return result

# Build and test character level chain
char_chain = build_char_markov(training_text.lower(), order=3)
char_output = generate_char_text(char_chain, "art", num_chars=200)

print("=" * 60)
print("🔤 CHARACTER LEVEL MARKOV CHAIN")
print("=" * 60)
print(f"\n📌 Seed: 'art'")
print(f"📝 Generated: {char_output}")
```

```python
print("=" * 60)
print("📊 COMPARING DIFFERENT MARKOV CHAIN ORDERS")
print("=" * 60)

for order in [1, 2, 3]:
    chain = build_markov_chain(training_text, order=order)

    # Get a valid seed
    first_state = list(chain.keys())[0]
    seed = list(first_state)

    generated = generate_text(chain, seed, num_words=25, order=order)
    print(f"\n🔹 Order {order}:")
    print(f"📝 {generated}")
    print("-" * 40)
```
